````python
import ast
import json
import glob
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

def safe_parse_prediction(x, debug_mode=False):
    """
    Parsing robusto per la colonna 'prediction'.
    Gestisce diversi formati: stringhe, liste gi√† parsate, JSON, PARSE_ERROR, etc.
    """
    if pd.isna(x) or x is None:
        if debug_mode:
            return [], "empty_or_nan"
        return []
    
    # Se √® gi√† una lista
    if isinstance(x, list):
        if debug_mode:
            return x, "already_list"
        return x
    
    # Se √® una stringa
    if isinstance(x, str):
        x = x.strip()
        if not x:
            if debug_mode:
                return [], "empty_string"
            return []
        
        # Gestisci valori di errore specifici
        if x in ['PARSE_ERROR', 'ERROR', 'FAILED', 'TIMEOUT', 'INVALID']:
            if debug_mode:
                return [], f"error_value_{x.lower()}"
            return []
        
        try:
            # Prova ast.literal_eval prima
            result = ast.literal_eval(x)
            if isinstance(result, list):
                if debug_mode:
                    return result, "ast_literal_eval_list"
                return result
            else:
                if debug_mode:
                    return [result], "ast_literal_eval_single"
                return [result]  # Wrap in lista se √® un singolo elemento
        except (ValueError, SyntaxError) as e:
            try:
                # Prova JSON parse
                result = json.loads(x)
                if isinstance(result, list):
                    if debug_mode:
                        return result, "json_loads_list"
                    return result
                else:
                    if debug_mode:
                        return [result], "json_loads_single"
                    return [result]
            except (json.JSONDecodeError, ValueError):
                # Se fallisce tutto, prova a splittare per virgole
                # Rimuovi caratteri come [ ] e ""
                cleaned = x.replace('[', '').replace(']', '').replace('"', '').replace("'", "")
                if ',' in cleaned:
                    result = [item.strip() for item in cleaned.split(',') if item.strip()]
                    if debug_mode:
                        return result, "comma_split"
                    return result
                else:
                    result = [cleaned] if cleaned else []
                    if debug_mode:
                        return result, "single_cleaned" if cleaned else "empty_after_clean"
                    return result
    
    # Fallback: converti in stringa e prova di nuovo
    if debug_mode:
        result, reason = safe_parse_prediction(str(x), debug_mode=True)
        return result, f"fallback_to_str_{reason}"
    return safe_parse_prediction(str(x))

def calculate_metrics(df):
    """
    Calcola le metriche di performance da un DataFrame.
    """
    if df.empty:
        return {
            'top1_accuracy': 0.0,
            'top5_hit_rate': 0.0,
            'mrr': 0.0,
            'catalogue_coverage': 0.0,
            'prediction_count': 0
        }
    
    # Normalizza le predizioni e ground truth
    def normalize_prediction_list(pred_list):
        if not isinstance(pred_list, list):
            return []
        return [str(item).strip() for item in pred_list if item and str(item).strip()]
    
    def normalize_ground_truth(gt):
        return str(gt).strip() if gt and str(gt).strip() != 'nan' else ""
    
    df_clean = df.copy()
    df_clean['prediction_norm'] = df_clean['prediction_list'].apply(normalize_prediction_list)
    df_clean['ground_truth_norm'] = df_clean['ground_truth'].apply(normalize_ground_truth)
    
    # Filtra righe con ground truth valido
    df_clean = df_clean[df_clean['ground_truth_norm'] != ""].copy()
    
    if df_clean.empty:
        return {
            'top1_accuracy': 0.0,
            'top5_hit_rate': 0.0,
            'mrr': 0.0,
            'catalogue_coverage': 0.0,
            'prediction_count': 0
        }
    
    # Calcola metriche
    def top1_accuracy(row):
        pred_norm = row['prediction_norm']
        if not pred_norm or len(pred_norm) == 0:
            return False
        return pred_norm[0] == row['ground_truth_norm']
    
    def top_k_hit(row, k=5):
        pred_norm = row['prediction_norm']
        if not pred_norm or len(pred_norm) == 0:
            return False
        return row['ground_truth_norm'] in pred_norm[:k]
    
    def reciprocal_rank(row, k=5):
        pred_norm = row['prediction_norm']
        if not pred_norm or len(pred_norm) == 0:
            return 0.0
        try:
            rank = pred_norm[:k].index(row['ground_truth_norm']) + 1
            return 1.0 / rank
        except ValueError:
            return 0.0
    
    # Applica calcoli
    df_clean['hit@1'] = df_clean.apply(top1_accuracy, axis=1)
    df_clean['hit@5'] = df_clean.apply(top_k_hit, axis=1)
    df_clean['rr'] = df_clean.apply(reciprocal_rank, axis=1)
    
    # Calcola coverage
    predicted_pois = set()
    for pred_list in df_clean['prediction_norm']:
        predicted_pois.update(pred_list)
    
    ground_truth_pois = set(df_clean['ground_truth_norm'].unique())
    ground_truth_pois.discard("")  # Rimuovi valori vuoti
    
    coverage = len(predicted_pois) / len(ground_truth_pois) if len(ground_truth_pois) > 0 else 0
    
    return {
        'top1_accuracy': df_clean['hit@1'].mean(),
        'top5_hit_rate': df_clean['hit@5'].mean(),
        'mrr': df_clean['rr'].mean(),
        'catalogue_coverage': coverage,
        'prediction_count': len(df_clean)
    }

def load_model_data(model_dir, model_name):
    """
    Carica tutti i CSV di un modello e calcola le metriche.
    Versione robusta con migliore gestione errori.
    """
    print(f"üìÇ Caricando dati per {model_name} da {model_dir}...")
    
    if not Path(model_dir).exists():
        print(f"   ‚ùå Directory non trovata: {model_dir}")
        return pd.DataFrame()
    
    csv_pattern = str(Path(model_dir) / "*_pred_*.csv")
    csv_files = glob.glob(csv_pattern)
    
    if not csv_files:
        print(f"   ‚ùå Nessun file CSV trovato in {model_dir}")
        return pd.DataFrame()
    
    print(f"   üìÑ Trovati {len(csv_files)} file CSV")
    
    # Statistiche globali per il debug
    total_loaded = 0
    total_before_filter = 0
    total_parse_errors = 0
    dfs = []
    failed_files = []
    
    for csv_file in csv_files:
        file_name = Path(csv_file).name
        try:
            # Caricamento con gestione errori
            try:
                df = pd.read_csv(csv_file)
            except pd.errors.ParserError:
                print(f"   ‚ö†Ô∏è  Parser error in {file_name}, utilizzando error handling...")
                df = pd.read_csv(csv_file, on_bad_lines='skip', engine='python')

            initial_rows = len(df)
            total_before_filter += initial_rows
            
            # Estrai l'anno dal nome file  
            year_token = next((part for part in Path(csv_file).stem.split('_')
                              if part.isdigit() and len(part) == 4), None)
            df['year'] = int(year_token) if year_token else np.nan

            # Parsing robusto delle predizioni CON GESTIONE ERRORI MIGLIORATA
            def safe_parse_with_fallback(pred_text):
                try:
                    return safe_parse_prediction(pred_text)
                except Exception as e:
                    # Se il parsing fallisce completamente, restituisce lista vuota
                    # invece di far crashare tutto il caricamento
                    return []
            
            df['prediction_list'] = df['prediction'].apply(safe_parse_with_fallback)
            
            # Conta gli errori di parsing in questo file
            parse_errors = sum(1 for pred in df['prediction'] if isinstance(pred, str) and pred.strip() in ['PARSE_ERROR', 'ERROR', 'FAILED', 'TIMEOUT', 'INVALID'])
            total_parse_errors += parse_errors
            
            # Verifica che le predizioni siano liste (INCLUDE anche liste vuote [])
            df_filtered = df[df['prediction_list'].apply(lambda x: isinstance(x, list))].copy()
            
            after_filter_rows = len(df_filtered)
            
            if after_filter_rows > 0:
                dfs.append(df_filtered)
                total_loaded += after_filter_rows
                success_rate = (after_filter_rows / initial_rows * 100) if initial_rows > 0 else 0
                
                if parse_errors > 0:
                    print(f"       {file_name}: {after_filter_rows:,} righe ({success_rate:.1f}%, {parse_errors} PARSE_ERROR)")
                else:
                    print(f"       {file_name}: {after_filter_rows:,} righe")
            else:
                failed_files.append(file_name)
                print(f"   ‚ö†Ô∏è  {file_name}: 0 righe utilizzabili dopo filtro")
            
        except Exception as e:
            failed_files.append(file_name)
            print(f"   ‚ùå Errore caricando {file_name}: {e}")
            continue
    
    if dfs:
        df_combined = pd.concat(dfs, ignore_index=True)
        print(f"   ‚úÖ {model_name}: {len(df_combined):,} righe totali")
        
        # Mostra statistiche dettagliate
        loss_pct = ((total_before_filter - total_loaded) / total_before_filter * 100) if total_before_filter > 0 else 0
        if loss_pct > 5 or total_parse_errors > 0:
            print(f"   üìä Statistiche: {total_parse_errors:,} PARSE_ERROR, {loss_pct:.1f}% righe perse")
        
        if failed_files:
            print(f"   ‚ö†Ô∏è  File falliti: {len(failed_files)}/{len(csv_files)}")
            
        return df_combined
    else:
        print(f"   ‚ùå Nessun file processato con successo per {model_name}")
        return pd.DataFrame()

print("‚úÖ Funzioni di caricamento dati robuste definite (con gestione PARSE_ERROR e calculate_metrics)")
````

### 3.3 Heat Map Performance

````python
# ============================================================================
# HEAT MAP PERFORMANCE NORMALIZZATA
# ============================================================================

if len(successfully_loaded) < 2:
    print("‚è≠Ô∏è  Saltando heat map: servono almeno 2 modelli")
else:
    print("üé® Creazione heat map performance normalizzata...")
    
    # Prepara dati per heat map
    metrics_for_heatmap = ['top1_accuracy', 'top5_hit_rate', 'mrr', 'catalogue_coverage']
    metric_labels = ['Top-1\nAccuracy', 'Top-5\nHit Rate', 'Mean Reciprocal\nRank', 'Catalogue\nCoverage']
    
    # Crea matrice dati
    heatmap_data = []
    model_labels = []
    
    for model_key in successfully_loaded:
        model_config = MODELS[model_key]
        metrics = model_metrics[model_key]
        
        row_data = []
        for metric in metrics_for_heatmap:
            if metric == 'catalogue_coverage':
                # Coverage pu√≤ essere > 1, quindi normalizziamo diversamente
                value = min(metrics[metric] * 100, 100)  # Cap a 100%
            else:
                value = metrics[metric] * 100
            row_data.append(value)
        
        heatmap_data.append(row_data)
        model_labels.append(model_config['name'])
    
    # Converte a numpy array per facilit√†
    heatmap_matrix = np.array(heatmap_data)
    
    # Crea heat map
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Usa colormap personalizzata
    im = ax.imshow(heatmap_matrix, cmap='RdYlGn', aspect='auto', vmin=0, vmax=100)
    
    # Configura assi
    ax.set_xticks(range(len(metric_labels)))
    ax.set_xticklabels(metric_labels, rotation=0, ha='center')
    ax.set_yticks(range(len(model_labels)))
    ax.set_yticklabels(model_labels)
    
    # Aggiungi valori nelle celle
    for i in range(len(model_labels)):
        for j in range(len(metric_labels)):
            value = heatmap_matrix[i, j]
            text_color = 'white' if value < 50 else 'black'
            ax.text(j, i, f'{value:.1f}%', ha='center', va='center', 
                   color=text_color, fontweight='bold', fontsize=10)
    
    ax.set_title('Performance Heat Map - Model Comparison\n(Normalized 0-100%)', 
                fontsize=16, fontweight='bold', pad=20)
    
    # Colorbar
    cbar = plt.colorbar(im, ax=ax, shrink=0.8, aspect=20)
    cbar.set_label('Performance (%)', rotation=270, labelpad=20, fontsize=12)
    
    plt.tight_layout()
    
    # Salva
    output_path = Path(OUTPUT_DIRS['comparison']) / 'models_performance_heatmap.png'
    plt.savefig(output_path, dpi=300, bbox_inches='tight', facecolor='white')
    print(f"üíæ Salvato: {output_path}")
    
    plt.show()
    
    # Analisi best/worst per categoria
    print("\nüìä ANALISI HEAT MAP:")
    print("-" * 40)
    
    for j, metric_name in enumerate(['Top-1 Accuracy', 'Top-5 Hit Rate', 'MRR', 'Coverage']):
        values_for_metric = heatmap_matrix[:, j]
        best_idx = np.argmax(values_for_metric)
        worst_idx = np.argmin(values_for_metric)
        
        print(f"{metric_name}:")
        print(f"  ü•á Best:  {model_labels[best_idx]} ({values_for_metric[best_idx]:.1f}%)")
        print(f"  ü•â Worst: {model_labels[worst_idx]} ({values_for_metric[worst_idx]:.1f}%)")
        print()
    
    print("‚úÖ Heat map performance completata!")
````

````python
# ============================================================================
# ANALISI TEMPORALE MULTI-MODELLO
# ============================================================================

if len(successfully_loaded) < 2:
    print("‚è≠Ô∏è  Saltando analisi temporale: servono almeno 2 modelli")
else:
    print("üìÖ Analisi temporale multi-modello...")
    
    # Verifica che abbiamo dati temporali
    temporal_data_available = False
    for model_key in successfully_loaded:
        df = model_data[model_key]
        if 'year' in df.columns and not df['year'].isna().all():
            temporal_data_available = True
            break
    
    if not temporal_data_available:
        print("   ‚ö†Ô∏è  Nessun dato temporale disponibile (colonna 'year' mancante o vuota)")
    else:
        print("   ‚úÖ Dati temporali trovati, creazione grafici...")
        
        # Calcola metriche per anno per ogni modello
        temporal_results = {}
        all_years = set()
        
        for model_key in successfully_loaded:
            df = model_data[model_key]
            if 'year' in df.columns:
                # Filtra anni validi
                df_with_year = df[df['year'].notna()]
                
                if len(df_with_year) > 0:
                    # Calcola metriche per anno
                    yearly_metrics = {}
                    for year in df_with_year['year'].unique():
                        year_data = df_with_year[df_with_year['year'] == year]
                        if len(year_data) > 0:
                            metrics = calculate_metrics(year_data)
                            yearly_metrics[int(year)] = metrics
                            all_years.add(int(year))
                    
                    temporal_results[model_key] = yearly_metrics
        
        if len(temporal_results) == 0:
            print("   ‚ö†Ô∏è  Nessun risultato temporale calcolabile")
        else:
            # Ordina gli anni
            sorted_years = sorted(all_years)
            print(f"   üìä Anni disponibili: {sorted_years}")
            
            # Crea grafici temporali
            metrics_to_plot = [
                ('top1_accuracy', 'Top-1 Accuracy (%)', 'temporal_top1_comparison.png'),
                ('top5_hit_rate', 'Top-5 Hit Rate (%)', 'temporal_top5_comparison.png'),
                ('mrr', 'Mean Reciprocal Rank (%)', 'temporal_mrr_comparison.png')
            ]
            
            for metric_key, ylabel, filename in metrics_to_plot:
                fig, ax = plt.subplots(figsize=(12, 6))
                
                # Plot linea per ogni modello
                for model_key in successfully_loaded:
                    if model_key in temporal_results:
                        model_config = MODELS[model_key]
                        
                        # Prepara dati per questo modello
                        x_data = []
                        y_data = []
                        
                        for year in sorted_years:
                            if year in temporal_results[model_key]:
                                x_data.append(year)
                                y_data.append(temporal_results[model_key][year][metric_key] * 100)
                        
                        if len(x_data) > 0:
                            ax.plot(x_data, y_data, color=model_config['color'], 
                                   marker='o', linewidth=2, markersize=6,
                                   label=model_config['name'])
                
                ax.set_xlabel('Year', fontsize=12)
                ax.set_ylabel(ylabel, fontsize=12)
                ax.set_title(f'{ylabel.replace(" (%)", "")} - Temporal Comparison', 
                           fontsize=14, fontweight='bold')
                ax.legend(loc='best')
                ax.grid(True, alpha=0.3)
                
                # Evidenzia periodo COVID se presente
                covid_years = [2020, 2021]
                if any(year in sorted_years for year in covid_years):
                    for covid_year in covid_years:
                        if covid_year in sorted_years:
                            ax.axvline(covid_year, color='red', linestyle='--', 
                                     alpha=0.5, linewidth=1)
                    ax.text(0.02, 0.98, 'Red lines: COVID-19 period', 
                           transform=ax.transAxes, fontsize=10, 
                           verticalalignment='top', 
                           bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
                
                plt.tight_layout()
                
                # Salva
                output_path = Path(OUTPUT_DIRS['comparison']) / filename
                plt.savefig(output_path, dpi=300, bbox_inches='tight', facecolor='white')
                print(f"üíæ Salvato: {output_path}")
                
                plt.show()
            
            # Analisi di stabilit√† temporale
            print("\nüìä ANALISI STABILIT√Ä TEMPORALE:")
            print("-" * 50)
            
            for model_key in successfully_loaded:
                if model_key in temporal_results and len(temporal_results[model_key]) > 1:
                    model_name = MODELS[model_key]['name']
                    
                    # Calcola variabilit√† per MRR
                    mrr_values = [temporal_results[model_key][year]['mrr'] * 100 
                                 for year in temporal_results[model_key].keys()]
                    mrr_std = np.std(mrr_values)
                    mrr_mean = np.mean(mrr_values)
                    stability_score = mrr_mean / (mrr_std + 1e-6)  # Higher = more stable
                    
                    print(f"{model_name:>15s}: MRR avg={mrr_mean:.1f}%, std={mrr_std:.1f}%, stability={stability_score:.1f}")
            
            print("\n‚úÖ Analisi temporale completata!")
````
